```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def show(pil_im):
    plt.axis('off')
    plt.imshow(np.asarray(pil_im))
    plt.margins(0,0)
    plt.show()
    

im=Image.open('a.jpg')
w,h=im.size

print(f'w:{w} h:{h}')
show(im)

pil2tensor=transforms.ToTensor()
tensor2pil=transforms.ToPILImage()
```

```python
p1=transforms.RandomCrop((h,w),padding=100)
for _ in range(5):
    im1=p1(im)
    show(im1)
```

```python
p2=transforms.RandomHorizontalFlip()
for _ in range(5):
    im2=p2(im)
    show(im2)
```

```python
p3=transforms.FiveCrop((h//2,w//2))
for im3 in p3(im):
    show(im3)
```

```python
p4=transforms.RandomErasing(p=0.5, scale=(0.02, 0.33), ratio=(0.3, 3.3), value=0)

for _ in range(5):
    im4=p4(pil2tensor(im))
    show(tensor2pil(im4))
```

```python
p5=transforms.ColorJitter(brightness=1, contrast=1, saturation=1, hue=0)
for _ in range(5):
    im5=p5(im)
    show(im5)
```

```python
p6=transforms.RandomAffine(degrees=10, translate=(0.2,0.2), scale=(0.5,1.5), shear=None, resample=False, fillcolor=0)
for _ in range(5):
    im6=p6(im)
    show(im6)
```
